```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata as ad
```

```python
with open("GSE60361_C1-3005-Expression.txt", 'r') as file:
    data = file.read().split()
    columns = [data[i:i+3006] for i in range(0, len(data), 3006)]
    df = pd.DataFrame({col[0]: col[1:] for col in columns[1:]})
df
```

```python
df = df.astype(int)
df.sum(axis=1)
```

```python
df.index = columns[0][1:]
df
adata = ad.AnnData(X = df.values,
                        var = pd.DataFrame(index = df.columns),
                        obs = pd.DataFrame(index = columns[0][1:]))
adata
```

```python
adata.write_h5ad("GSE60361_processed.h5ad")
adata
```
